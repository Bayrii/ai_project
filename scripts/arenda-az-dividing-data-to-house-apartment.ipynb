Load core analysis libraries and set pandas display options for clearer inspection output.

## Load Raw Property Dataset and Quick Shape Check
Read the source CSV and print row and column counts for an initial sanity check.

```python
df = pd.read_csv('../data/satilir_properties.csv')
print("Data loaded successfully!")
# - Dataset shape (hint: use df.shape)
print("Dataset shape:", df.shape)
# - Number of observations (hint: df.shape[0] gives rows)
print("Number of observations:", df.shape[0])
# - Number of features (hint: df.shape[1] gives columns)
print("Number of features:", df.shape[1])
```

## Drop Unused Columns and Remove Extreme Area Outlier
Remove non-essential fields and filter the known anomalous area_m2 value before splitting.

```python
df.drop(columns=['link', 'temirli'], inplace=True)
df.drop(df[df['area_m2'] == 1600].index, inplace=True)
df.head()
```

## Split Listings into Apartment and House DataFrames
Use land_area_sot availability to separate apartment records from house or villa records.

```python
df_apartment = df.loc[df['land_area_sot'].isna()]
df_apartment.drop(columns=['land_area_sot'], inplace=True)
df_house = df.loc[~df['land_area_sot'].isna()]
```

## Save Split Datasets to Disk
Export apartment and house DataFrames to separate CSV files in their target folders.

```python
df_apartment.to_csv('../data/apartment/satilir_properties_apartment.csv', index=False)
df_house.to_csv('../data/house/satilir_properties_house.csv', index=False)

print("Apartment dataset saved to 'data/apartment/satilir_properties_apartment.csv'") 
print("House dataset saved to 'data/house/satilir_properties_house.csv'") 
```
